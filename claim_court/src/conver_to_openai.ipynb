````python
import pandas as pd
````

````python
instruction_test1 = """
You are a fact-checking judge. You will receive facts from a Supreme Court case and a claim based on the Supreme Court case. Your task is to determine whether the claim is factually consistent with the information from the case.

Decision categories (choose only one)
Supported: The claim is explicitly stated in, or clearly supported by, the Supreme Court case evidence.
Contradicted: The Supreme Court case evidence contains information that directly conflicts with the claim.

Rules
1. Base your judgment only on the Supreme Court evidence.
2. If the evidence does not support the claim, you must not label it as SUPPORTED.
3. Do not rely on outside knowledge or assumptions.
4. Do not invent information that is not in the evidence.

Evidence:{{evidence}}
# Output Format:
Return your judgement and explanation in a JSON format as follows:
```json
{{
"explanation": "the reasoning behind the judgement",
"judgement": "<Supported|Contradicted>",
}}
```
"""
````

````python
input_path="/home/qxw5305/projects/claim_court/data/claim_with_fact.csv"
````

````python
#inputs = pd.read_json(input_path, lines=True)
#inputs.text.values
inputs = pd.read_csv(input_path)

inputs['facts']
````

````python

openai_format = [
    {
        'custom_id': f"claim_{i}",
        'method': 'POST',
        'url': '/v1/chat/completions',
        'body': {
            'messages': [
                {
                    'role': 'system',
                    'content': instruction_test1.format(evidence=fact)
                },
                {
                    'role': 'user',
                    'content': claim
                }
            ],
            'model': 'google/gemma-3-27b-it'
        }
    }
    for i, (claim, fact) in enumerate(
        zip(inputs['claim'], inputs['facts']+inputs['api_question']+inputs['api_conclusion'])
    )
]

````

````python
pd.DataFrame(openai_format).to_json('prompts_1.jsonl', orient='records', lines=True)
````
