```python
import sqlite3
import pandas as pd
import os
import ast
from sentence_transformers import SentenceTransformer, util
import torch
from tqdm import tqdm

# 1. Load Bills from Database
def load_bills(db_path):
    print(f"Loading bills from {db_path}...")
    conn = sqlite3.connect(db_path)
    
    # Select relevant columns
    # Constructing ID as: "{BillCongress} {BillType} {BillNumber}"
    query = """
    SELECT 
        BillCongress, 
        BillType, 
        BillNumber, 
        BillDescription, 
        BillSummary, 
        BillSubjects 
    FROM Bills
    """
    
    df = pd.read_sql_query(query, conn)
    conn.close()
    
    print(f"Loaded {len(df)} bills.")
    
    # Construct ID and Text
    # Text = Title/Description + Summary + Subjects
    
    ids = []
    texts = []
    
    for idx, row in tqdm(df.iterrows(), total=len(df), desc="Processing Bills"):
        # ID Construction
        # Ensure integers are formatted as integers
        try:
            congress = int(row['BillCongress'])
            number = int(row['BillNumber'])
            bill_type = row['BillType']
            bill_id = f"{congress} {bill_type} {number}"
        except ValueError:
            # Fallback if conversion fails
            bill_id = f"{row['BillCongress']} {row['BillType']} {row['BillNumber']}"
            
        ids.append(bill_id)
        
        # Text Construction
        text_parts = []
        if row['BillDescription']:
            text_parts.append(str(row['BillDescription']))
        if row['BillSummary']:
            text_parts.append(str(row['BillSummary']))
        if row['BillSubjects']:
            text_parts.append(str(row['BillSubjects']))
            
        full_text = " ".join(text_parts)
        texts.append(full_text)
        
    return ids, texts

# 2. Load Claims from CSV
def load_claims(csv_path):
    print(f"Loading claims from {csv_path}...")
    df = pd.read_csv(csv_path)
    
    claims = []
    ground_truth = [] # List of sets of bill IDs
    
    for idx, row in df.iterrows():
        claim_text = row['claim_entered']
        sourced_bills_raw = row['sourced_bills']
        
        try:
            # Parse the list string "['ID1', 'ID2']"
            if isinstance(sourced_bills_raw, str):
                bill_ids = ast.literal_eval(sourced_bills_raw)
            else:
                bill_ids = []
                
            if claim_text and bill_ids:
                claims.append(claim_text)
                ground_truth.append(set(bill_ids))
                
        except Exception as e:
            print(f"Error parsing row {idx}: {e}")
            
    print(f"Loaded {len(claims)} valid claims.")
    return claims, ground_truth

# Paths
db_path = '../data_store/claimlens/congress.db'
claims_path = '../data_store/claimlens/vote-claims.csv'

# Load Data
bill_ids, bill_texts = load_bills(db_path)
claims, ground_truth = load_claims(claims_path)

# Create a mapping for quick lookup if needed, though for search we just need the lists
# We need to ensure bill_ids align with bill_texts for the corpus
print(f"First 3 Bill IDs: {bill_ids[:3]}")
print(f"First 3 Bill Texts: {bill_texts[:3]}")
```

```python
def evaluate_model(model_name_or_path, corpus_texts, corpus_ids, query_texts, ground_truth, k_values=[1, 5, 10, 50]):
    print(f"Evaluating model: {model_name_or_path}")
    
    # Load model
    try:
        model = SentenceTransformer(model_name_or_path, trust_remote_code=True)
    except Exception as e:
        print(f"Failed to load model {model_name_or_path}: {e}")
        return None

    # Encode Corpus
    print("Encoding corpus...")
    # Using a smaller batch size to be safe with memory, though 32 is usually fine
    corpus_embeddings = model.encode(corpus_texts, batch_size=32, show_progress_bar=True, convert_to_tensor=True)
    
    # Encode Queries
    print("Encoding queries...")
    query_embeddings = model.encode(query_texts, batch_size=32, show_progress_bar=True, convert_to_tensor=True)
    
    # Search
    print("Searching...")
    # We want top K results
    max_k = max(k_values)
    results = util.semantic_search(query_embeddings, corpus_embeddings, top_k=max_k)
    
    # Calculate Metrics
    metrics = {f"Recall@{k}": 0.0 for k in k_values}
    
    for query_idx, hits in enumerate(results):
        gold_ids = ground_truth[query_idx]
        
        # Get retrieved IDs
        retrieved_ids = [corpus_ids[hit['corpus_id']] for hit in hits]
        
        for k in k_values:
            top_k_retrieved = set(retrieved_ids[:k])
            
            # Check if ANY relevant bill is retrieved
            intersection = len(top_k_retrieved.intersection(gold_ids))
            recall = intersection / len(gold_ids) if len(gold_ids) > 0 else 0
            metrics[f"Recall@{k}"] += recall
            
    # Average
    for k in k_values:
        metrics[f"Recall@{k}"] /= len(query_texts)
        
    return metrics

# Define Models
models_to_test = {
    "BAAI/bge-base-en-v1.5": ["../store/model_testing/baai_1", "../store/model_testing/baai_3", "../store/model_testing/baai_5"],
    "sentence-transformers/all-MiniLM-L6-v2": ["../store/model_testing/minilm_1", "../store/model_testing/minilm_3", "../store/model_testing/minilm_5"],
    "Qwen/Qwen3-Embedding-0.6B": ["../store/model_testing/qwen_0.6_1", "../store/model_testing/qwen_0.6_3"]
}

# Run Evaluation
results_summary = []

for base_model, trained_models in models_to_test.items():
    print(f"\n--- Processing Base Model: {base_model} ---")
    
    # Evaluate Base
    base_metrics = evaluate_model(base_model, bill_texts, bill_ids, claims, ground_truth)
    if base_metrics:
        print(f"Base Model Results: {base_metrics}")
        results_summary.append({"model": base_model, "type": "base", **base_metrics})
    
    # Evaluate Trained
    for trained_model_path in trained_models:
        if os.path.exists(trained_model_path):
            trained_metrics = evaluate_model(trained_model_path, bill_texts, bill_ids, claims, ground_truth)
            if trained_metrics:
                print(f"Trained Model ({os.path.basename(trained_model_path)}) Results: {trained_metrics}")
                results_summary.append({"model": os.path.basename(trained_model_path), "type": "trained", "base": base_model, **trained_metrics})
        else:
            print(f"Trained model path not found: {trained_model_path}")

# Display Summary
print("\n=== Final Summary ===")
df_results = pd.DataFrame(results_summary)
display(df_results)
```
