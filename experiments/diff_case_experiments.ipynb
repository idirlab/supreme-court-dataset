```python
import pandas as pd

# Load the CSV files
overlap_df = pd.read_csv('../results/overlap_diff_case.csv')
contradiction_df = pd.read_csv('../results/contradiction_diff_case.csv')

# Print the head of both dataframes
print("Overlap Diff Case Head:")
display(overlap_df.head())
print("\nContradiction Diff Case Head:")
display(contradiction_df.head())
```

```python
print("Overlap Decision Counts:")
print(overlap_df['decision'].value_counts())
print("\nContradiction Decision Counts:")
print(contradiction_df['decision'].value_counts())
```

```python
# Filter and print examples for Overlap (Redundant)
redundant_examples = overlap_df[overlap_df['decision'] == 'redundant'].head(10)
print("--- Redundant Examples (Overlap) ---")
for i, row in redundant_examples.iterrows():
    print(f"Example {i+1}:")
    print(f"Case 1: {row['name1']}")
    print(f"Claim 1: {row['claim1']}")
    print(f"Case 2: {row['name2']}")
    print(f"Claim 2: {row['claim2']}")
    print("-" * 20)

# Filter and print examples for Contradiction (Contradiction)
contradiction_examples = contradiction_df[contradiction_df['decision'] == 'contradiction'].head(10)
print("\n--- Contradiction Examples (Contradiction) ---")
for i, row in contradiction_examples.iterrows():
    print(f"Example {i+1}:")
    print(f"Case 1: {row['name1']}")
    print(f"Claim 1: {row['claim1']}")
    print(f"Case 2: {row['name2']}")
    print(f"Claim 2: {row['claim2']}")
    print("-" * 20)
```
