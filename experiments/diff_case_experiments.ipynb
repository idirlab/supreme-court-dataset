```python
import pandas as pd

# Load resolved contradictions
resolved_contra_path = '../results/factual_claims_resolved_contradictions.csv'
if pd.io.common.file_exists(resolved_contra_path):
    df_resolved = pd.read_csv(resolved_contra_path)

    # Load metadata for dispositions
    metadata_path = '../clean_data_with_details.csv'
    df_meta = pd.read_csv(metadata_path)

    # Helper to normalize dispositions
    def normalize_disp(d):
        return str(d).lower().strip()

    df_meta['disposition_norm'] = df_meta['disposition'].apply(normalize_disp)

    # Allowed dispositions for the *overruling* case (the one that wins)
    # If Case A overrules Case B, Case A must have one of these dispositions.
    allowed_dispositions = [
        'reversed/remanded', 
        'reversed', 
        'reversed in-part/remanded', 
        'reversed in-part'
    ]
    allowed_dispositions = [normalize_disp(x) for x in allowed_dispositions]

    # Merge dispositions
    # Case 1
    df_merged = df_resolved.merge(
        df_meta[['docket', 'name', 'disposition_norm']],
        left_on=['docket1', 'name1'],
        right_on=['docket', 'name'],
        how='left'
    )
    df_merged.rename(columns={'disposition_norm': 'disposition1'}, inplace=True)
    df_merged.drop(columns=['docket', 'name'], inplace=True)

    # Case 2
    df_merged = df_merged.merge(
        df_meta[['docket', 'name', 'disposition_norm']],
        left_on=['docket2', 'name2'],
        right_on=['docket', 'name'],
        how='left'
    )
    df_merged.rename(columns={'disposition_norm': 'disposition2'}, inplace=True)
    df_merged.drop(columns=['docket', 'name'], inplace=True)

    # Apply Logic
    def process_row(row):
        decision = str(row.get('decision', '')).strip()
        disp1 = row.get('disposition1', '')
        disp2 = row.get('disposition2', '')
        
        final_action = "keep_both" # Default (consistent or invalid overrule)
        
        if decision == 'case1_overruled':
            # Case 2 overrules Case 1. Case 2 must be a reversal.
            if disp2 in allowed_dispositions:
                final_action = "drop_claim1"
            else:
                final_action = "keep_both_invalid_disp"
                
        elif decision == 'case2_overruled':
            # Case 1 overrules Case 2. Case 1 must be a reversal.
            if disp1 in allowed_dispositions:
                final_action = "drop_claim2"
            else:
                final_action = "keep_both_invalid_disp"
                
        elif decision == 'consistent':
            final_action = "keep_both_consistent"
            
        return final_action

    df_merged['final_action'] = df_merged.apply(process_row, axis=1)

    # Stats
    print("Final Action Counts:")
    print(df_merged['final_action'].value_counts())

    print("\nDetails:")
    print(f"Total pairs: {len(df_merged)}")
    print(f"Valid Overrules (Drop 1): {len(df_merged[df_merged['final_action'] == 'drop_claim1'])}")
    print(f"Valid Overrules (Drop 2): {len(df_merged[df_merged['final_action'] == 'drop_claim2'])}")
    print(f"Invalid Overrules (Disposition mismatch): {len(df_merged[df_merged['final_action'] == 'keep_both_invalid_disp'])}")
    
    # Display some invalid overrules to check
    print("\n--- Examples of Invalid Overrules (Disposition Mismatch) ---")
    invalid_examples = df_merged[df_merged['final_action'] == 'keep_both_invalid_disp'].head(5)
    for i, row in invalid_examples.iterrows():
        print(f"Decision: {row['decision']}")
        print(f"Case 1 ({row['name1']}): {row['disposition1']}")
        print(f"Case 2 ({row['name2']}): {row['disposition2']}")
        print(f"Explanation: {row['explanation']}")
        print("-" * 20)

    # Save the results
    output_path = '../results/diff_case_removed_contra.csv'
    df_merged.to_csv(output_path, index=False)
    print(f"\nSaved results to {output_path}")
else:
    print(f"File not found: {resolved_contra_path}")
```

```python
import pandas as pd
import networkx as nx
import json
import os

# 1. Load Data
overlaps_path = '../results/factual_claims_resolved_overlaps_diff_case.csv'
contradictions_path = '../results/factual_claims_resolved_contradictions_diff_case.csv'
filtered_claims_path = '../filtered_claims.csv'
metadata_path = '../clean_data_with_details.csv'

# Check if filtered_claims exists, else try factual_claims
if not os.path.exists(filtered_claims_path):
    print(f"{filtered_claims_path} not found, trying ../factual_claims.csv")
    filtered_claims_path = '../factual_claims.csv'

df_overlaps = pd.read_csv(overlaps_path)
df_contradictions = pd.read_csv(contradictions_path)
df_claims = pd.read_csv(filtered_claims_path)
df_meta = pd.read_csv(metadata_path)

print(f"Loaded {len(df_claims)} filtered claims.")
print(f"Loaded {len(df_overlaps)} overlap pairs.")
print(f"Loaded {len(df_contradictions)} contradiction pairs.")

# Normalize dispositions for checking
def normalize_disp(d):
    return str(d).lower().strip()

df_meta['disposition_norm'] = df_meta['disposition'].apply(normalize_disp)
disposition_map = dict(zip(df_meta['name'], df_meta['disposition_norm']))

allowed_dispositions = [
    'reversed/remanded', 
    'reversed', 
    'reversed in-part/remanded', 
    'reversed in-part'
]
allowed_dispositions = [normalize_disp(x) for x in allowed_dispositions]

# Create lookup for claims
# Assuming (name, claim) is unique. 
df_claims['claim_clean'] = df_claims['claim'].astype(str).str.strip()

# Identify the name column (could be 'name' or 'case_name')
name_col = 'name'
if 'case_name' in df_claims.columns:
    name_col = 'case_name'
elif 'name' not in df_claims.columns:
    print("Warning: Neither 'name' nor 'case_name' found in claims file. Using first column?")
    name_col = df_claims.columns[0]

print(f"Using '{name_col}' as the case name column.")
df_claims['name_clean'] = df_claims[name_col].astype(str).str.strip()
df_claims['id'] = df_claims.index

# Map (name, claim) -> ID
claim_lookup = {}
for idx, row in df_claims.iterrows():
    claim_lookup[(row['name_clean'], row['claim_clean'])] = idx

def get_claim_id(name, claim):
    return claim_lookup.get((str(name).strip(), str(claim).strip()))

print("Lookup map created.")
```

```python
# 2. Process Overlaps (Merge Conflicts)
G_overlap = nx.Graph()
G_overlap.add_nodes_from(df_claims['id'])

overlap_conflicts = 0
for _, row in df_overlaps.iterrows():
    id1 = get_claim_id(row['name1'], row['claim1'])
    id2 = get_claim_id(row['name2'], row['claim2'])
    
    if id1 is not None and id2 is not None:
        decision = str(row.get('resolution_decision', '')).strip()
        if decision in ['keep_1', 'keep_2', 'merge']:
            G_overlap.add_edge(id1, id2, decision=decision, merged_claim=row.get('merged_claim'))
            
            if G_overlap.degree[id1] > 1 or G_overlap.degree[id2] > 1:
                overlap_conflicts += 1

print(f"Overlap Merge Conflicts (nodes involved in multiple pairs): {overlap_conflicts}")

# Find components
raw_components = list(nx.connected_components(G_overlap))
print(f"Found {len(raw_components)} raw components (including singletons).")

# Refine components (Split large ones > 5)
components = []
for comp in raw_components:
    if len(comp) <= 5:
        components.append(comp)
    else:
        # Split large component: Merge along sub-branches with different cases
        # Strategy: Greedy clustering ensuring unique cases in each cluster
        subgraph = G_overlap.subgraph(comp)
        remaining_nodes = set(comp)
        
        while remaining_nodes:
            # Pick a start node
            start_node = next(iter(remaining_nodes))
            remaining_nodes.remove(start_node)
            
            cluster = {start_node}
            # Track cases in this cluster to ensure uniqueness
            cluster_cases = {df_claims.loc[start_node, name_col]}
            
            # BFS/Traversal to grow cluster
            queue = [start_node]
            
            while queue:
                u = queue.pop(0)
                for v in subgraph.neighbors(u):
                    if v in remaining_nodes:
                        v_case = df_claims.loc[v, name_col]
                        # Only merge if the case is not already in the cluster
                        if v_case not in cluster_cases:
                            cluster.add(v)
                            cluster_cases.add(v_case)
                            remaining_nodes.remove(v)
                            queue.append(v)
            
            components.append(cluster)

print(f"Refined to {len(components)} components after splitting large ones (>5).")

multi_node_components = [c for c in components if len(c) > 1]
print(f"Found {len(multi_node_components)} multi-claim components to merge.")

# Example of merge conflict
print("\n--- Examples of Merges ---")
count = 0
for c in multi_node_components:
    if len(c) > 2:
        print(f"Complex Merge (Size {len(c)}): Merging cases for these claims:")
        # Print names
        for node_id in list(c):
            print(f" - Case: {df_claims.loc[node_id, name_col]} | Claim: {df_claims.loc[node_id, 'claim'][:50]}...")
        print("-" * 20)
        count += 1
        if count >= 3: break

if count == 0 and len(multi_node_components) > 0:
    print("No complex merges (>2) found. Showing simple merge:")
    c = multi_node_components[0]
    for node_id in list(c):
        print(f" - Case: {df_claims.loc[node_id, name_col]} | Claim: {df_claims.loc[node_id, 'claim'][:50]}...")
```

```python
# 3. Process Contradictions
# Filter valid contradictions based on disposition
valid_contradictions = []

for _, row in df_contradictions.iterrows():
    id1 = get_claim_id(row['name1'], row['claim1'])
    id2 = get_claim_id(row['name2'], row['claim2'])
    
    if id1 is not None and id2 is not None:
        decision = str(row.get('decision', '')).strip()
        disp1 = disposition_map.get(str(row['name1']).strip(), '')
        disp2 = disposition_map.get(str(row['name2']).strip(), '')
        
        winner = None
        loser = None
        
        if decision == 'case1_overruled':
            # Case 2 overrules Case 1. Case 2 must be reversal.
            if disp2 in allowed_dispositions:
                winner = id2
                loser = id1
        elif decision == 'case2_overruled':
            # Case 1 overrules Case 2. Case 1 must be reversal.
            if disp1 in allowed_dispositions:
                winner = id1
                loser = id2
        
        if winner is not None:
            valid_contradictions.append({
                'winner': winner,
                'loser': loser,
                'winner_name': row['name2'] if winner == id2 else row['name1'],
                'loser_name': row['name1'] if winner == id2 else row['name2']
            })

print(f"Found {len(valid_contradictions)} valid overruling pairs.")

# Check for multiple overrulings (A overruled by B, A overruled by C)
loser_counts = {}
for item in valid_contradictions:
    l = item['loser']
    loser_counts[l] = loser_counts.get(l, 0) + 1

multi_overruled = {k: v for k, v in loser_counts.items() if v > 1}
print(f"Claims overruled multiple times (by different cases): {len(multi_overruled)}")
if len(multi_overruled) > 0:
    print("Example IDs of multi-overruled claims:", list(multi_overruled.keys())[:5])
    # Print details for one
    example_id = list(multi_overruled.keys())[0]
    print(f"Example Claim ID {example_id} overruled by:")
    for item in valid_contradictions:
        if item['loser'] == example_id:
            print(f" - {item['winner_name']}")

# Check for serial overrulings (A overrules B, B overrules C)
winners = set(item['winner'] for item in valid_contradictions)
losers = set(item['loser'] for item in valid_contradictions)
serial = winners.intersection(losers)
print(f"Serial Overrulings (A->B->C): {len(serial)}")
if len(serial) > 0:
    print("Example IDs in middle of chain:", list(serial)[:5])
```

```python
# 4. Apply Updates and Create Final DataFrame

# Map original ID -> Super Claim ID (Leader of component)
id_map = {}
component_data = {} # leader_id -> {ids: [], text: ..., cases: []}

for comp in components:
    comp_list = list(comp)
    leader = comp_list[0] # Pick first as leader
    
    # Determine text
    # If any merge decision in component, try to find merged text
    final_text = df_claims.loc[leader, 'claim']
    
    # Check edges for merged text
    subgraph = G_overlap.subgraph(comp)
    for u, v, data in subgraph.edges(data=True):
        if data.get('decision') == 'merge':
            mc = data.get('merged_claim')
            if mc and isinstance(mc, str) and len(mc) > 10:
                final_text = mc
                break
    
    # Collect cases from ALL nodes in the component
    cases = []
    for node in comp:
        id_map[node] = leader
        # Use the identified name column
        cases.append(df_claims.loc[node, name_col])
    
    component_data[leader] = {
        'ids': comp_list,
        'text': final_text,
        'cases': sorted(list(set(cases))), # Unique cases
        'type': 'confirmed' if len(comp) > 1 else 'none',
        'overruling_case': None
    }

# Apply Contradictions
claims_to_drop = set()

for item in valid_contradictions:
    # Find the leader for the winner and loser (in case they were part of overlaps)
    # Note: id_map contains all nodes because components include singletons
    winner_leader = id_map[item['winner']]
    loser_leader = id_map[item['loser']]
    
    if winner_leader == loser_leader:
        # Contradiction within the same merged component? Should be rare/impossible if logic holds.
        continue
        
    # Update Winner
    winner_data = component_data[winner_leader]
    loser_data = component_data[loser_leader]
    
    # Merge cases (Winner takes Loser's cases into its history)
    winner_data['cases'].extend(loser_data['cases'])
    winner_data['cases'] = sorted(list(set(winner_data['cases'])))
    
    # Update type and overruling info
    winner_data['type'] = 'overruling'
    # If already overruled, maybe append? For now, just take the last one or keep list.
    # User asked for "overruling_case" column.
    if winner_data['overruling_case'] is None:
        winner_data['overruling_case'] = item['winner_name']
    else:
        # If multiple overrulers, maybe list them?
        if item['winner_name'] not in winner_data['overruling_case']:
             winner_data['overruling_case'] += f"; {item['winner_name']}"
    
    # Mark loser for drop
    claims_to_drop.add(loser_leader)

# Construct Final DataFrame
final_rows = []
for leader, data in component_data.items():
    if leader in claims_to_drop:
        continue
        
    # Get original row data
    row = df_claims.loc[leader].copy()
    
    # Update fields
    row['claim'] = data['text']
    row['case_name'] = json.dumps(data['cases']) # JSON structure
    row['type'] = data['type']
    row['overruling_case'] = data['overruling_case']
    
    # Remove helper cols
    if 'claim_clean' in row: del row['claim_clean']
    if 'name_clean' in row: del row['name_clean']
    if 'id' in row: del row['id']
    # Remove original 'name' if we are using 'case_name'
    if 'name' in row: del row['name']
    
    final_rows.append(row)

df_final = pd.DataFrame(final_rows)

# Ensure 'claim' is first
cols = ['claim'] + [c for c in df_final.columns if c != 'claim']
df_final = df_final[cols]

print(f"Final DataFrame Size: {len(df_final)}")
print(df_final['type'].value_counts())

# Save
output_path = '../diff_case_claims_updated.csv'
df_final.to_csv(output_path, index=False)
print(f"Saved to {output_path}")
```

```python
import pandas as pd
import json
import os

output_path = '../diff_case_claims_updated.csv'

if os.path.exists(output_path):
    df_final = pd.read_csv(output_path)
    print(f"Loaded {len(df_final)} rows.")
    
    # Check for duplicates
    dup_count = df_final.duplicated(subset=['claim']).sum()
    print(f"Duplicate claims found: {dup_count}")
    
    if dup_count > 0:
        print("Merging duplicates...")
        rows = []
        for claim, group in df_final.groupby('claim'):
            if len(group) == 1:
                rows.append(group.iloc[0])
            else:
                # Base row
                row = group.iloc[0].copy()
                
                # Merge cases
                all_cases = set()
                for c_json in group['case_name']:
                    try:
                        cases = json.loads(c_json)
                        if isinstance(cases, list):
                            all_cases.update(cases)
                    except:
                        pass
                row['case_name'] = json.dumps(sorted(list(all_cases)))
                
                # Merge type
                types = set(group['type'])
                if 'overruling' in types:
                    row['type'] = 'overruling'
                elif 'confirmed' in types:
                    row['type'] = 'confirmed'
                
                # Merge overruling cases
                ocs = set()
                for oc in group['overruling_case'].dropna():
                    ocs.add(oc)
                if ocs:
                    row['overruling_case'] = "; ".join(sorted(list(ocs)))
                
                rows.append(row)
        
        df_final = pd.DataFrame(rows)
        print(f"New size: {len(df_final)}")
        df_final.to_csv(output_path, index=False)
    
    print("\nHead of Final Dataset:")
    display(df_final.head())
    
    print("\nType Distribution:")
    print(df_final['type'].value_counts())

else:
    print("File not found.")
```

```python
import pandas as pd
import networkx as nx
import json
import os

# 1. Load Data
overlaps_path = '../results/factual_claims_resolved_overlaps_diff_case.csv'
contradictions_path = '../results/factual_claims_resolved_contradictions_diff_case.csv'
filtered_claims_path = '../filtered_claims.csv'
metadata_path = '../clean_data_with_details.csv'

# Check if filtered_claims exists, else try factual_claims
if not os.path.exists(filtered_claims_path):
    print(f"{filtered_claims_path} not found, trying ../factual_claims.csv")
    filtered_claims_path = '../factual_claims.csv'

df_overlaps = pd.read_csv(overlaps_path)
df_contradictions = pd.read_csv(contradictions_path)
df_claims = pd.read_csv(filtered_claims_path)
df_meta = pd.read_csv(metadata_path)

print(f"Loaded {len(df_claims)} filtered claims.")
print(f"Loaded {len(df_overlaps)} overlap pairs.")
print(f"Loaded {len(df_contradictions)} contradiction pairs.")

# Normalize dispositions for checking
def normalize_disp(d):
    return str(d).lower().strip()

df_meta['disposition_norm'] = df_meta['disposition'].apply(normalize_disp)
disposition_map = dict(zip(df_meta['name'], df_meta['disposition_norm']))

allowed_dispositions = [
    'reversed/remanded', 
    'reversed', 
    'reversed in-part/remanded', 
    'reversed in-part'
]
allowed_dispositions = [normalize_disp(x) for x in allowed_dispositions]

# Create lookup for claims
df_claims['claim_clean'] = df_claims['claim'].astype(str).str.strip()

# Identify the name column
name_col = 'name'
if 'case_name' in df_claims.columns:
    name_col = 'case_name'
elif 'name' not in df_claims.columns:
    name_col = df_claims.columns[0]

print(f"Using '{name_col}' as the case name column.")
df_claims['name_clean'] = df_claims[name_col].astype(str).str.strip()
df_claims['id'] = df_claims.index

# Map (name, claim) -> ID
claim_lookup = {}
for idx, row in df_claims.iterrows():
    claim_lookup[(row['name_clean'], row['claim_clean'])] = idx

def get_claim_id(name, claim):
    return claim_lookup.get((str(name).strip(), str(claim).strip()))

# 2. Process Overlaps (Merge Conflicts) - Reusing logic to build components
G_overlap = nx.Graph()
G_overlap.add_nodes_from(df_claims['id'])

for _, row in df_overlaps.iterrows():
    id1 = get_claim_id(row['name1'], row['claim1'])
    id2 = get_claim_id(row['name2'], row['claim2'])
    
    if id1 is not None and id2 is not None:
        decision = str(row.get('resolution_decision', '')).strip()
        if decision in ['keep_1', 'keep_2', 'merge']:
            G_overlap.add_edge(id1, id2, decision=decision, merged_claim=row.get('merged_claim'))

# Find components
raw_components = list(nx.connected_components(G_overlap))

# Refine components (Split large ones > 5)
components = []
for comp in raw_components:
    if len(comp) <= 5:
        components.append(comp)
    else:
        subgraph = G_overlap.subgraph(comp)
        remaining_nodes = set(comp)
        
        while remaining_nodes:
            start_node = next(iter(remaining_nodes))
            remaining_nodes.remove(start_node)
            
            cluster = {start_node}
            cluster_cases = {df_claims.loc[start_node, name_col]}
            
            queue = [start_node]
            
            while queue:
                u = queue.pop(0)
                for v in subgraph.neighbors(u):
                    if v in remaining_nodes:
                        v_case = df_claims.loc[v, name_col]
                        if v_case not in cluster_cases:
                            cluster.add(v)
                            cluster_cases.add(v_case)
                            remaining_nodes.remove(v)
                            queue.append(v)
            
            components.append(cluster)

print(f"Refined to {len(components)} components.")

# 3. Initialize Component Data
id_map = {}
component_data = {} # leader_id -> {ids: [], text: ..., cases: []}

for comp in components:
    comp_list = list(comp)
    leader = comp_list[0]
    
    final_text = df_claims.loc[leader, 'claim']
    
    subgraph = G_overlap.subgraph(comp)
    for u, v, data in subgraph.edges(data=True):
        if data.get('decision') == 'merge':
            mc = data.get('merged_claim')
            if mc and isinstance(mc, str) and len(mc) > 10:
                final_text = mc
                break
    
    cases = []
    for node in comp:
        id_map[node] = leader
        cases.append(df_claims.loc[node, name_col])
    
    component_data[leader] = {
        'ids': comp_list,
        'text': final_text,
        'cases': sorted(list(set(cases))),
        'type': 'confirmed' if len(comp) > 1 else 'none',
        'overruling_case': None
    }

# 4. Process Contradictions (UPDATED LOGIC)
valid_contradictions = []

for _, row in df_contradictions.iterrows():
    id1 = get_claim_id(row['name1'], row['claim1'])
    id2 = get_claim_id(row['name2'], row['claim2'])
    
    if id1 is not None and id2 is not None:
        decision = str(row.get('decision', '')).strip()
        disp1 = disposition_map.get(str(row['name1']).strip(), '')
        disp2 = disposition_map.get(str(row['name2']).strip(), '')
        
        winner = None
        loser = None
        
        if decision == 'case1_overruled':
            if disp2 in allowed_dispositions:
                winner = id2
                loser = id1
        elif decision == 'case2_overruled':
            if disp1 in allowed_dispositions:
                winner = id1
                loser = id2
        
        if winner is not None:
            # Find leaders
            winner_leader = id_map[winner]
            loser_leader = id_map[loser]
            
            if winner_leader != loser_leader:
                winner_name = row['name2'] if winner == id2 else row['name1']
                
                # Update Loser (The overruled claim)
                loser_data = component_data[loser_leader]
                loser_data['type'] = 'overruled'
                
                if loser_data['overruling_case'] is None:
                    loser_data['overruling_case'] = winner_name
                else:
                    if winner_name not in loser_data['overruling_case']:
                         loser_data['overruling_case'] += f"; {winner_name}"

# 5. Construct Final DataFrame
final_rows = []
for leader, data in component_data.items():
    # We keep ALL claims now (no claims_to_drop)
    
    row = df_claims.loc[leader].copy()
    
    row['claim'] = data['text']
    row['case_name'] = json.dumps(data['cases'])
    row['type'] = data['type']
    row['overruling_case'] = data['overruling_case']
    
    if 'claim_clean' in row: del row['claim_clean']
    if 'name_clean' in row: del row['name_clean']
    if 'id' in row: del row['id']
    if 'name' in row: del row['name']
    
    final_rows.append(row)

df_final = pd.DataFrame(final_rows)

# Ensure 'claim' is first
cols = ['claim'] + [c for c in df_final.columns if c != 'claim']
df_final = df_final[cols]

print(f"Final DataFrame Size: {len(df_final)}")
print(df_final['type'].value_counts())

# Save
output_path = '../diff_case_claims_updated.csv'
df_final.to_csv(output_path, index=False)
print(f"Saved to {output_path}")
```

```python
import pandas as pd
import json
import os
import numpy as np

# 1. Load Data
current_file_path = '../diff_case_claims_with_negatives.csv'
contradictions_path = '../results/factual_claims_resolved_contradictions_diff_case.csv'
metadata_path = '../clean_data_with_details.csv'

print("Loading files...")
df_current = pd.read_csv(current_file_path)
df_contradictions = pd.read_csv(contradictions_path)
df_meta = pd.read_csv(metadata_path)

print(f"Current dataset shape: {df_current.shape}")
print(f"Contradictions shape: {df_contradictions.shape}")

# 2. Setup Disposition Logic (to identify Winner/Loser correctly)
def normalize_disp(d):
    return str(d).lower().strip()

df_meta['disposition_norm'] = df_meta['disposition'].apply(normalize_disp)
disposition_map = dict(zip(df_meta['name'], df_meta['disposition_norm']))

allowed_dispositions = [
    'reversed/remanded', 
    'reversed', 
    'reversed in-part/remanded', 
    'reversed in-part'
]
allowed_dispositions = [normalize_disp(x) for x in allowed_dispositions]

# 3. Identify Valid Contradictions and Losers
# We need to find the losers that correspond to the winners present in df_current.

# First, identify all valid (Winner, Loser, Loser_Claim) tuples from contradictions
valid_contradictions = []

for _, row in df_contradictions.iterrows():
    decision = str(row.get('decision', '')).strip()
    disp1 = disposition_map.get(str(row['name1']).strip(), '')
    disp2 = disposition_map.get(str(row['name2']).strip(), '')
    
    winner_name = None
    loser_name = None
    loser_claim = None
    
    if decision == 'case1_overruled':
        # Case 2 overrules Case 1. Case 2 must be reversal.
        if disp2 in allowed_dispositions:
            winner_name = row['name2']
            loser_name = row['name1']
            loser_claim = row['claim1']
    elif decision == 'case2_overruled':
        # Case 1 overrules Case 2. Case 1 must be reversal.
        if disp1 in allowed_dispositions:
            winner_name = row['name1']
            loser_name = row['name2']
            loser_claim = row['claim2']
    
    if winner_name and loser_name:
        valid_contradictions.append({
            'winner': winner_name,
            'loser': loser_name,
            'loser_claim': loser_claim
        })

print(f"Found {len(valid_contradictions)} valid contradiction pairs in source data.")

# 4. Identify Winners present in the current dataset
# The 'overruling_case' column in df_current (for type='overruling') contains the Winner's name.
overruling_rows_mask = df_current['type'] == 'overruling'
overruling_rows = df_current[overruling_rows_mask]

print(f"Found {len(overruling_rows)} rows with type='overruling' in current dataset.")

# Extract all winner names referenced in the dataset
# The 'overruling_case' column might contain multiple names separated by "; "
present_winners = set()
for val in overruling_rows['overruling_case'].dropna():
    names = [n.strip() for n in str(val).split(';')]
    present_winners.update(names)

print(f"Identified {len(present_winners)} unique winner cases referenced in dataset.")

# 5. Filter Contradictions to find relevant Losers
# We only care about losers overruled by the winners we found
relevant_losers = []
for item in valid_contradictions:
    if item['winner'] in present_winners:
        relevant_losers.append(item)

print(f"Found {len(relevant_losers)} relevant dropped claims (losers) to restore.")

# 6. Create Rows for Dropped Claims
# Group by (loser_name, loser_claim) to handle multiple overrulers
loser_map = {} # (name, claim) -> set of winners

for item in relevant_losers:
    key = (item['loser'], item['loser_claim'])
    if key not in loser_map:
        loser_map[key] = set()
    loser_map[key].add(item['winner'])

new_rows = []
next_fact_id = df_current['fact_id'].max() + 1 if 'fact_id' in df_current.columns else 0

for (l_name, l_claim), winners in loser_map.items():
    # Create new row
    new_row = {
        'claim': l_claim,
        'case_name': json.dumps([l_name]),
        'judgement': np.nan, # Missing
        'explanation': np.nan, # Missing
        'raw_output': np.nan, # Missing
        'type': 'overruled',
        'overruling_case': "; ".join(sorted(list(winners))),
        'fact_id': next_fact_id,
        'label': 1, # Assuming positive sample
        'is_negation': False
    }
    new_rows.append(new_row)
    next_fact_id += 1

print(f"Created {len(new_rows)} new rows for overruled claims.")

# 7. Update Current Dataset
# Change type 'overruling' -> 'none' and clear 'overruling_case'
# Note: We do this for ALL 'overruling' rows, as they are now just normal claims (winners)
df_current.loc[overruling_rows_mask, 'type'] = 'none'
df_current.loc[overruling_rows_mask, 'overruling_case'] = np.nan

# 8. Append New Rows
if new_rows:
    df_new = pd.DataFrame(new_rows)
    df_final = pd.concat([df_current, df_new], ignore_index=True)
else:
    df_final = df_current

print(f"Final dataset shape: {df_final.shape}")
print("Type distribution:")
print(df_final['type'].value_counts())

# 9. Save
output_path = '../diff_case_claims_with_negatives_fixed.csv'
df_final.to_csv(output_path, index=False)
print(f"Saved fixed dataset to {output_path}")

# Optional: Overwrite original if desired, but saving to new file first for safety
# df_final.to_csv(current_file_path, index=False)
```

```python
import pandas as pd
import os

# Load the fixed dataset
file_path = '../diff_case_claims_with_negatives_fixed.csv'

if os.path.exists(file_path):
    df = pd.read_csv(file_path)
    
    # 1. Drop invalid negations (overruled + is_negation=True)
    # These shouldn't exist based on our logic, but good to filter
    mask_drop = (df['type'] == 'overruled') & (df['is_negation'] == True)
    df_clean = df[~mask_drop]
    
    # 2. Identify missing negations
    # Find fact_ids that already have a negation
    has_negation = df_clean[df_clean['is_negation'] == True]['fact_id'].unique()
    
    # Filter for candidates:
    # - Not in has_negation (so we need to generate one)
    # - type != 'overruled' (we don't generate negations for overruled claims)
    # - is_negation == False (we look at the positive sample to generate the negative)
    candidates = df_clean[
        (~df_clean['fact_id'].isin(has_negation)) & 
        (df_clean['type'] != 'overruled') & 
        (df_clean['is_negation'] == False)
    ]
    
    print(f"Number of claims requiring negation generation: {len(candidates)}")
else:
    print(f"File not found: {file_path}")
```
